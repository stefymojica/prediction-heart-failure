```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('../data/train_heart_failure.csv')
```

```python
X = df.drop(['target'], axis=1)
X
```

```python
y = df['target'].to_frame()
y
```

```python
obj_escaler = StandardScaler()
X_estandard = obj_escaler.fit_transform(X)
```

```python
# train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_estandard, y, stratify=y,test_size=0.2,random_state=1)
```

```python
model = MLPClassifier(hidden_layer_sizes=(10,2), activation='relu', random_state=10)
```

```python
model.fit(X_train,y_train)
```

```python
# score model
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
# load test dataset
df_test_target = pd.read_csv('../data/test_heart_failure.csv')
df_test_target.head()
```

```python
# sample prediction
y_in = np.asarray([67,1,1,4,1,2,1,1]).reshape(1,-1)
```

```python
# model prediction
predicts = model.predict(y_in)
predicts[0]
```
